```python
#TOTAL SALES IN SPECIFIC REGION
```

```python
cur=con.cursor()
select_stmt="SELECT REGIONAL_OFFICE_NAME,SUM(f.TOTAL_QUANTITY) as TOTAL_QUANTITY "\
"FROM snow.FACT_TABLE as f "\
"inner JOIN snow.LOCATION as L ON L.LOCATION_KEY = f.LOCATION_KEY "\
"GROUP BY CUBE(L.REGIONAL_OFFICE_NAME)"\
"ORDER BY SUM (f.TOTAL_QUANTITY) DESC"

cur.execute(select_stmt)
records=cur.fetchall()
records
```

```python
region_sales_df=pd.DataFrame(list(records),columns=["REGIONAL_OFFICE_NAME","TOTAL_QUANTITY"])
region_sales_df
```

```python
region_sales_df=region_sales_df.dropna()
region_sales_df
```

```python
#Which region sells the most quantity of items monthwise?
```

```python
con = PostgresConnection().getConnection()
cur=con.cursor()
select_stmt="SELECT REGIONAL_OFFICE_NAME,SUM(f.TOTAL_QUANTITY) as TOTAL_QUANTITY,T.MONTH "\
"FROM snow.FACT_TABLE as f "\
"inner JOIN snow.LOCATION as L ON L.LOCATION_KEY = f.LOCATION_KEY "\
"inner JOIN snow.TIME as T ON T.TIME_KEY = f.TIME_KEY "\
"GROUP BY CUBE(L.REGIONAL_OFFICE_NAME,T.MONTH)"\
"ORDER BY T.MONTH "

cur.execute(select_stmt)
records=cur.fetchall()
records
```

```python
region_sales_m_df=pd.DataFrame(list(records),columns=["REGIONAL_OFFICE_NAME","TOTAL_QUANTITY","MONTH"])
region_sales_m_df
```

```python
region_sales_m_df=region_sales_m_df.dropna()
region_sales_m_df
```
