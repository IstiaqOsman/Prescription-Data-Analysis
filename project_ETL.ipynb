### Generating Key 

##### Time Key

```python
dataDump.loc[0, 'YEAR_MONTH']
```

```python
TIME_KEY = []
for i in range(0,len(dataDump_df)):
    TIME_KEY.append('T' + str(dataDump_df._get_value(i, 'YEAR_MONTH')))

dataDump_df['TIME_KEY'] = pd.Series(TIME_KEY, index=dataDump_df.index)
dataDump_df
```

##### Location Key

```python
LOCATION_KEY = []
for i in range(1,len(dataDump_df)+1):
    if(i<10):
        LOCATION_KEY.append('L000000'+str(i))
    elif(10<=i and i<100):
        LOCATION_KEY.append('L00000'+str(i))
    elif(100<=i and i<1000):
       LOCATION_KEY.append('L0000'+str(i))
    elif(1000<=i and i<10000):
       LOCATION_KEY.append('L000'+str(i))
    elif(10000<=i and i<100000):
       LOCATION_KEY.append('L00'+str(i))
    else:
       LOCATION_KEY.append('L0'+str(i))
dataDump_df['LOCATION_KEY'] = pd.Series(LOCATION_KEY, index=dataDump_df.index)
dataDump_df
```

```python
string = str('L0' + dataDump._get_value(0, 'REGIONAL_OFFICE_CODE')[0] + dataDump._get_value(0, 'STP_CODE')[0] + 
    dataDump._get_value(0, 'PCO_CODE')[0] + dataDump._get_value(0, 'POSTCODE')[0:3])
string
```

```python
LOCATION_KEY2 = []
for i in range(0,len(dataDump_df)):
    LOCATION_KEY2.append(str('L0' + dataDump_df._get_value(i, 'REGIONAL_OFFICE_CODE') + dataDump_df._get_value(i, 'STP_CODE') + 
    dataDump_df._get_value(i, 'PCO_CODE') + dataDump_df._get_value(i, 'POSTCODE')))
dataDump_df['LOCATION_KEY2'] = pd.Series(LOCATION_KEY2, index=dataDump_df.index)
dataDump_df
```

##### Prescription Key 

```python
PRESCRIPTION_KEY = []
for i in range(1,len(dataDump_df)+1):
    if(i<10):
        PRESCRIPTION_KEY.append('P000000'+str(i))
    elif(10<=i and i<100):
        PRESCRIPTION_KEY.append('P00000'+str(i))
    elif(100<=i and i<1000):
       PRESCRIPTION_KEY.append('P0000'+str(i))
    elif(1000<=i and i<10000):
       PRESCRIPTION_KEY.append('P000'+str(i))
    elif(10000<=i and i<100000):
       PRESCRIPTION_KEY.append('P00'+str(i))
    else:
       PRESCRIPTION_KEY.append('P0'+str(i))
dataDump_df['PRESCRIPTION_KEY'] = pd.Series(PRESCRIPTION_KEY, index=dataDump_df.index)
dataDump_df
```
