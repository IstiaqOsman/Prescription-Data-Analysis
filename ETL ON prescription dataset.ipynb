```python
TIME_KEY = []
for i in range(1,len(data)+1):
    if(i<10):
        TIME_KEY.append('T00000'+str(i))
    elif(10<=i and i<100):
       TIME_KEY.append('T0000'+str(i))
    elif(100<=i and i<1000):
       TIME_KEY.append('T000'+str(i))
    elif(1000<=i and i<10000):
       TIME_KEY.append('T00'+str(i))
    elif(10000<=i and i<100000):
       TIME_KEY.append('T0'+str(i))
    else:
       TIME_KEY.append('T'+str(i))
data['TIME_KEY'] = pd.Series(TIME_KEY, index=data.index)
data
```

```python
LOCATION_KEY = []
for i in range(1,len(data)+1):
    if(i<10):
        LOCATION_KEY.append('L00000'+str(i))
    elif(10<=i and i<100):
        LOCATION_KEY.append('L0000'+str(i))
    elif(100<=i and i<1000):
       LOCATION_KEY.append('L000'+str(i))
    elif(1000<=i and i<10000):
       LOCATION_KEY.append('L00'+str(i))
    elif(10000<=i and i<100000):
       LOCATION_KEY.append('L0'+str(i))
    else:
       LOCATION_KEY.append('L'+str(i))
data['LOCATION_KEY'] = pd.Series(LOCATION_KEY, index=data.index)
data
```

```python
PRESCRIPTION_KEY = []
for i in range(1,len(data)+1):
    if(i<10):
        PRESCRIPTION_KEY.append('P00000'+str(i))
    elif(10<=i and i<100):
        PRESCRIPTION_KEY.append('P0000'+str(i))
    elif(100<=i and i<1000):
       PRESCRIPTION_KEY.append('P000'+str(i))
    elif(1000<=i and i<10000):
       PRESCRIPTION_KEY.append('P00'+str(i))
    elif(10000<=i and i<100000):
       PRESCRIPTION_KEY.append('P0'+str(i))
    else:
       PRESCRIPTION_KEY.append('P'+str(i))
data['PRESCRIPTION_KEY'] = pd.Series(PRESCRIPTION_KEY, index=data.index)
data
```
